```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Define the MLP model
def create_mlp_model(input_dim, hidden_units, output_dim):
    model = keras.Sequential()
    
    # Input layer
    model.add(layers.Input(shape=(input_dim,)))

    # Hidden layers
    for units in hidden_units:
        model.add(layers.Dense(units, activation='relu'))
    
    # Output layer
    model.add(layers.Dense(output_dim, activation='sigmoid'))

    return model

# Define your dataset and parameters
input_dim = 784  # Number of input features
hidden_units = [64, 32]  # Number of units in each hidden layer
output_dim = 10  # Number of output units (binary classification)

# Create the MLP model
model = create_mlp_model(input_dim, hidden_units, output_dim)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print a summary of the model architecture
model.summary()

# Training data and labels (replace with your data)
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape((-1, 28 * 28)).astype('float32') / 255.0
x_test = x_test.reshape((-1, 28 * 28)).astype('float32') / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=64)

# Evaluating the model with test da:
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}")
```
